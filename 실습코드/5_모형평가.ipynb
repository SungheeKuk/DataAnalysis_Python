## [프로그래밍][1차수행평가]: 메뉴 선택 키오스크 만들기

**학습 목표:**

1. `print()`를 사용하여 원하는 문자열과 숫자를 출력할 수 있다.
2. `input()`을 사용하여 사용자로부터 값을 입력받을 수 있다.
3. 변수를 만들어 숫자(가격, 수량)와 문자열(메뉴 이름)을 저장할 수 있다.
4. 입력받은 문자열을 `int()`를 사용해 숫자로 변환할 수 있다.
5. 덧셈(`+`)과 곱셈(`*`) 연산자를 사용하여 총금액을 계산할 수 있다.
6. **조건문(`if`, `elif`, `else`)을 사용하여 사용자의 선택에 따라 다른 코드를 실행할 수 있다.**

우리 학교 축제에서 사용할 간단한 음료 주문 키오스크를 만들려고 합니다.
이 키오스크는 두 가지 메뉴 중 **사용자가 선택한 딱 한 가지 메뉴**의 주문만 받습니다.

* **메뉴 1:** `아메리카노` (가격: 3000원)
* **메뉴 2:** `카페라떼` (가격: 3500원)

프로그램이 시작되면, 사용자에게 **(1) 아메리카노, (2) 카페라떼** 중 하나를 선택하게 합니다.
그다음, **선택한 메뉴의 수량**을 물어보고, **총 결제 금액**을 계산하여 영수증처럼 출력해야 합니다.

2. 아메리카노와 카페라떼의 가격(3000, 3500)을 각각 `price1`, `price2` 변수에 저장하세요.
3. 총 금액을 저장할 `total_price` 변수를 0으로 미리 만들어 두세요.
4. `input()`을 사용해 "메뉴 번호를 선택하세요 (1 또는 2): "라고 물어보고, 입력받은 값을 `choice` 변수에 저장하세요.
5. 입력받은 `choice`는 문자열이므로, `int()`를 사용해 숫자형으로 변환하여 `choice_int`에 저장하세요.
6. **`if` 조건문**을 사용하여 `choice_int`가 1인지 확인하세요.
   * 만약 1이라면, "아메리카노 몇 잔?"을 물어보고 수량을 입력받아 `count` 변수에 저장하세요.
   * `count`를 `int()`로 변환한 뒤, `total_price = price1 * (변환된 수량)` 공식을 사용해 총 금액을 계산하세요.
7. **`elif` 조건문**을 사용하여 `choice_int`가 2인지 확인하세요.
   * 만약 2이라면, "카페라떼 몇 잔?"을 물어보고 수량을 입력받아 `count` 변수에 저장하세요.
   * `count`를 `int()`로 변환한 뒤, `total_price = price2 * (변환된 수량)` 공식을 사용해 총 금액을 계산하세요.
8. **`else`를** 사용하여 1 또는 2가 아닌 다른 숫자를 입력했을 때 "잘못된 번호입니다."라고 출력하게 하세요.
9. 마지막으로, 계산된 `total_price`를 `print()`를 사용하여 보기 좋게 출력하세요. (단, 0원일 때는 출력하지 않아도 좋습니다.)
