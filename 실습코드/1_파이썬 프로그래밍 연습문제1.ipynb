###Q1
홍길동 씨의 주민등록번호는 881120-1068234이다. 홍길동씨의 주민등록번호를 연월일(YYYYMMDD)부분과 그 뒤의 숫자 부분으로 나누어 출력하여라

```python
number = "881120-1068234"
birth  = number[0:6]
secret = number[-7:]

print(birth)
print(secret)
```

## Q2
[‘Life’, ‘is’, ‘too’, ‘short’]라는 리스트를 Life is too short라는 문자열로 만들어 출력하여라.

```python
ex_ls     = ['Life', 'is', 'too', 'short']
sentence1 = ex_ls[0] + " " + ex_ls[1] + " " + ex_ls[2] + " " + ex_ls[3]
sentence2 = " ".join(ex_ls)

print(sentence1)
print(sentence2)
```

## Q3
(1,2,3)이라는 튜플에 4라는 값을 추가하여 (1,2,3,4)처럼 만들어 출력하여라.


```python
bs_tp = (1, 2, 3)
ad_tp = (4,)

res_tp = bs_tp + ad_tp
print(res_tp)
```
