## 1.1 숫자형
- 숫자형(Number)이란 숫자 형태로 이루어진 자료형으로, 이미 잘 알고 있는 것들 입니다. 숫자형으로 사용되는 항목에는 정수, 실수, 8진수, 16진수, 복소수 형태가 있습니다.
- 그 중, 정수형과 실수형의 형태가 가장 많이 활용되는 형태라고 할 수 있습니다.

### 1.1.3. 8진수와 16진수

```python
# 8진수는 주로 리눅스 권한 코드로 많이 사용됨
# 8진수 정의
# 숫자 앞에 0o 또는 0O을 붙여서 사용함
a = 0o177
b = 0O177
print(a)
print(b)

# 10진수를 8진수로 표현하는 방법
print(oct(8))
print(oct(123))
```

```python
# 16진수는 주로 색상코드에서 많이 사용됨
# 16진수 정의
# 숫자 앞에 0x 또는 0X를 붙여서 사용함
a = 0x8ff
b = 0XABC

print(a)
print(b)

print(hex(125))
print(hex(8))
```

### 1.1.4. 복소수

```python
# 복소수는 과학 공학 계산, 신호처리, 전기 회로 해석 등에서 많이 사용됨
a = 1+2j
b = 3-4J
# 복소수는 괄호로 표기됩니다.
print(a)
print(b)
```

```python
# 복소수의 실수 부분을 리턴
a.real
```

```python
# 복소수의 허수 부분을 리턴
a.imag
```

```python
# 복소수의 켤레복소수를 리턴
a.conjugate()
```

```python
# 복소수의 절대값을 리턴
print(abs(a))
a = 1+2j
b = 3-4J
print((a.real**2 + a.imag**2) ** 0.5)
```

- 문자열(String)이란 문자, 단어 등으로 구성된 문자들의 집합을 의미합니다.
- 문자열의 덧셈과 곱셈 연산으로부터 문자열을 수정할 수 있으며, 문자열 내에 특정 단어나 글자를 선택할 수 있습니다.

### 1.2.1. 문자열 입력방법

```python
# 양쪽에 "를 사용하는 경우
"Hello World"
```

```python
# 양쪽에 '를 사용하는 경우
'Python is fun'
```

```python
# 양쪽에 """를 사용하는 경우
"""Life is too short, You need python"""
```

```python
# 양쪽에 '''를 사용하는 경우
'''Life is too short, You need python'''
```

```python
# 양쪽에 '를 사용하는 경우는 보통 작성하는 문장 안에 큰 따옴표(")가 포함되어야 하는 경우입니다.
'"Python is very easy." he says.'
```

```python
# 양쪽에 '''나, """를 사용하는 경우는 보통 작성하는 문장이 여러 줄로 이루어진 경우에 사용됩니다.
multiline = '''
Life is too short
You need python
'''
print(multiline)
```

```python
multiline2 = """
Life is too short
You need python
"""
print(multiline2)
```

```python
# 만약 """이나, '''을 사용하지 않은 경우에는 \n를 사용합니다.
multiline3 = "Life is too short \nYou need python"
print(multiline3)
```

### 2.3.1. 딕셔너리

```python
# 딕셔너리 정의 리스트들
dic.items()
```

### 2.4.1. 집합
- 집합은 파이썬2.3부터 지원되기 시작한 자료형으로 집합에 관련된 것들을 쉽게 처리하기 위해 만들어진 자료형 입니다.


```python
# 집합 정의
s1 = set([1,2,3])
s1
```

```python
# 문자열도 가능
s2 = set("Hello")
s2
```

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
```

```python
# 교집합
print(s1 & s2)
print(s1.intersection(s2))
```

```python
# 합집합
print(s1 | s2)
print(s1.union(s2))

s3 = set([1,2,3,4,5,6])
print(s1 | s3)
```

```python
# 차집합
print(s1 - s2)
print(s1.difference(s2))
```

```python
# 집합 추가 (하나의 요소)
s1 = set([1,2,3])
s1.add(4)
s1
```

```python
# update를 통한 여러개의 집합 추가
s1.update([4,5,6])
s1
```

```python
# 문자열을 추가하는 경우 각 문자들을 분리하여 추가하게 됨
s1.update("Python")
print(s1)
# 하나의 문자열로 추가하고싶은 경우?
s1.add("Python")
print(s1)
```

```python
# 요소 삭제
s1.remove(2)
s1
```
